```python
import pandas as pd
import altair as alt

data = pd.read_excel('data representing - weather.xlsx')
data.head()
```

```python
chart = alt.Chart(data).mark_circle(size=100).encode(
    x='WeatherValue',
    y='ActualValue',
    color='Weather Type',
    tooltip=['Day', 'Actual', 'Weather Type']
).interactive().properties(
    title='Relationship Between Weather and Mood (Feb 2026)'
)

chart
```

```python
chart2 = alt.Chart(data).mark_line(point=True).encode(
    x='Day',
    y='ActualValue',
    color='Weather Type',
    tooltip=['Day', 'Actual']
).interactive().properties(
    title='Mood Changes Over Time'
)

chart2
```

```python
chart3 = alt.Chart(data).mark_bar().encode(
    x='Weather Type',
    y='mean(ActualValue)',
    color='Weather Type'
).properties(
    title='Average Mood by Weather Type'
)

chart3
```
